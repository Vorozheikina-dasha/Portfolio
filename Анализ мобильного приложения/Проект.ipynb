## Исследовательский анализ

На этом шаге мы анализируем датасет. Как ведут себя пользователи в приложении. В данных отсутсвуют логи сессий, поэтому мы их выделим сами.

### Рассчет Ratention Rate (RR)

```python
#Для начала объединим таблицы
df = mobile_dataset.merge(mobile_sourсes, on='user_id', how='left')
df
```

#### Выделение пользовательских сессий

Пользователи могут заходить в приложение с разной частотой. Чаще всего пользователи "залипают" в поисках вещи, пока есть свободное время: в дороге, за обедом, в перерывах и тд. Поэтому время сессии логично взять в 30 минут, как так один пользователь может зайти утром и вечером в приложение.

```python
#Сортируем датасет
df = df.sort_values(['user_id', 'event_time'])

#Возвращаем разность 
diff_timestamp = df.groupby('user_id')['event_time'].diff()

#Вводим таймауд
timeout = timedelta(minutes=30)

new_session = (diff_timestamp.isnull()) | (diff_timestamp > timeout)

#Создаем id для уникальных сессий
df['session_id'] = df.loc[new_session, ['user_id', 'event_time']] \
    .groupby('user_id').rank(method='first').astype(int)

#Уберем пропуски
df['session_id'] = df['session_id'].fillna(method='ffill').astype(int)
df.head(5)
```

```python
#Выделим время начала сессий по пользователю
session_start = df.groupby(['user_id', 'session_id']).agg({'event_time': 'min'})
session_start = session_start.rename(columns={'event_time': 'session_start'})
session_start.head(5)
```

```python
#Выделим время первого заходав приложение по пользователю
session_first = df.groupby('user_id').agg({'event_time': 'min'})
session_first = session_first.rename(columns={'event_time': 'first_ts'})
session_first.head(5)
```

```python
#Объединим таблицы
sessions = session_start.merge(session_first, on='user_id', how='left').reset_index()
sessions.head(5)
```

#### Расчет RR

```python
#Вычисляем лайфтайм для каждой сессии в днях

sessions['lifetime'] = (
    sessions['session_start'] - sessions['first_ts']
).dt.days

sessions.head(5)
```

```python
#Строим таблицу удержания

sessions['dt'] = sessions['first_ts'].dt.date

retention_grop = sessions.pivot_table(
    index=['dt'], columns='lifetime', values='user_id', aggfunc='nunique'
)

retention_grop
```

```python
#Вычисляем размеры когорт

cohort_sizes = (
    sessions.groupby('dt')
    .agg({'user_id': 'nunique'})
    .rename(columns={'user_id': 'cohort_size'})
)

cohort_sizes
```

```python
#Объединяем размеры когорт и таблицу удержания

retention_rate = cohort_sizes.merge(
    retention_grop, on='dt', how='left'
).fillna(0)

retention_rate
```

```python
#Делим данные таблицы удержания на размеры когорт

retention_rate = retention_rate.div(
    retention_rate['cohort_size'], axis=0
).drop(columns=['cohort_size'])

retention_rate
```

```python
#Строим хитмэп

plt.figure(figsize=(25, 15)) 
sns.heatmap(
    retention_rate.drop(columns=[0]),
    annot=True,
    fmt='.2%'
)
plt.title('Тепловая карта удержания')
plt.xlabel('Lifetime')

plt.show()
```

Пользователи, пришедшие 21 октября показали самый высокий процент удержания на второй день (9.32%). На третий день самый высокий процент удержания у пользователей, пришедших 10 октября (8.05%).

Так же пользователи от 21 октябля на 8 день удержания показали процентр равный второму дню. 

В основонм вся активность пользователей проходит в первую неделю(+-день) использования приложения, далее интерес сокращается. Причем пользователи наиболее активны в первые 3 дня, затем идет спад, а через неделю они возвращаются снова. Это может быть связано как с тем, что они находят вещи в перые 2-3 дня, а далее просто скролят объявления. Либо с тем, что пропадает интерес к поиску, если за первые дни в приложении не найдены нужные вещи.

### Расчет времени, которое пользователи проводят в приложении

#### Подготовка данных

```python
#Добавляем столбец с окончанием сессии
session_end = df.groupby(['user_id', 'session_id']).agg({'event_time': 'max'})
session_end = session_end.rename(columns={'event_time': 'session_end'})

#Добавляем столбец с длительностью сессии
session_duration = session_start.merge(session_end, on=['user_id', 'session_id'], how='left')
session_duration['duration'] = session_duration['session_end'] - session_duration['session_start']

#Считаем сессии в секундах
session_duration['duration_s'] = session_duration['duration'].apply(lambda x: x.total_seconds())

session_duration['dt'] = session_duration['session_start'].dt.date

#Добавляем столбец с днем недели
session_duration['weekday'] = session_duration['session_start'].dt.weekday

#Добавляем столбец с часом
session_duration['hour'] = session_duration['session_start'].dt.hour

session_duration.reset_index()

#Соединим данные
full_df = df.merge(session_start, on=['user_id', 'session_id'], how='left')
full_df = full_df.merge(session_duration, on=['user_id', 'session_start'], how='left')
full_df
```

```python
full_df.duplicated().sum()
```

```python
#Удалим появившиеся дубликаты
full_df = full_df.drop_duplicates().reset_index(drop=True)
```

#### Расчет времени

```python
#Количество пользователей заходящих в приложение
def visits_hist(column, bins, figsize, part_title, x_label):
    full_df[column].hist(bins=bins, figsize=figsize)
    plt.title(part_title + ' посещения пользователями приложение', fontsize = 25)
    plt.xlabel(x_label)
    plt.ylabel('Количество посещений')

    return plt.show()
```

```python
visits_hist('session_start', 28, (25,15), 'Даты', 'Даты посещения')
```

## Сегментация пользователей

В рамках этого шага мы выделяем несколько сегментов пользователей, чтобы в дальнейшем оптимизировать приложение. Будет выбран основной сегмент и смежные.

На основе анализа данных, мы можем выделить следующие 4 сегмента: 
- пользователи, которые заходят ночью в 2-3
- пользователи, которые заходят утром в 11
- пользователи, которые заходят днем в 14-15 часов
- пользователи, которые заходят вечером в 19 часов

Такие сегменты выбраны на основе роста активности у пользователей в приложении по времени использования и количеству сессий.

Такие сегмента дадут понять, в какое время пользователи вовлечены в поиск вещей, а когда "коротают" время.

Предположительно основным сегментом могуть стать пользователи, которые заходят вечером в 19 часов. У него будет конверсия выше остальных. Вслед за ним будет сегмент пользователей, которые заходят 15 часов. Прверим.

### Какой сегмент лучше по по RR

```python
def rr(df, text):
    
    #Выделим время первого захода в приложение по пользователю
    f_df = df.groupby('user_id').agg({'event_time': 'min'})
    f_df = f_df.rename(columns={'event_time': 'first_ts'})
    
    #Объединим таблицы
    data = df.merge(f_df, on='user_id', how='left')
    
    #Рассчитываем lifetime 
    data['lifetime'] = (
    data['session_start'] - data['first_ts']
    ).dt.days
    
    data['dt'] = data['first_ts'].dt.date

    #Строим таблицу удержания
    retention_grop = data.pivot_table(
    index=['dt'], columns='lifetime', values='user_id', aggfunc='nunique'
    )
    
    #Вычисляем размеры когорт
    cohort_sizes = (
        data.groupby('dt')
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'cohort_size'})
    )
    
    #Объединяем размеры когорт и таблицу удержания
    retention_rate = cohort_sizes.merge(
        retention_grop, on='dt', how='left'
    ).fillna(0)
    
    #Делим данные таблицы удержания на размеры когорт
    retention_rate = retention_rate.div(
        retention_rate['cohort_size'], axis=0
    ).drop(columns=['cohort_size'])
    
    #Строим тепловую карту
    plt.figure(figsize=(25, 15)) 
    sns.heatmap(
        retention_rate.drop(columns=[0]),
        annot=True,
        fmt='.2%'
    )
    
    plt.title('Тепловая карта удержания ' + text)
    plt.xlabel('Lifetime')

    return plt.show()
```

## Проверка статистических гипотез

### Проверка гипотезы, которую выдвинул заказчик

Некоторые пользователи установили приложение по ссылке из yandex, другие — из google. 

**Гипотеза:**  две эти группы демонстрируют разную конверсию в просмотры контактов.

Для начала сформулируем нулевую и альтернативную гипотезы.

H0. Конверсия пользователей в просмотры контактов, пришедших из Yandex равна конверсии пользователей, пришедших из Google.

Н1. Конверсия пользователей в просмотры контактов, пришедших из Yandex не равна конверсии пользователей, пришедших из Google.

Уровень значимости поставим 0,05.

```python
#Посчитаем конверсии
def z_test(data1, data2):
    
    s1 = len(data1.query('event_name == "show_contacts"')['user_id'].unique())  
    s2 = len(data2.query('event_name == "show_contacts"')['user_id'].unique()) 

    u1 = len(data1['user_id'].unique())
    u2 = len(data2['user_id'].unique())

    p1 = s1/u1
    p2 = s2/u2
    p = (s1 + s2) / (u1 + u2)
    se = np.sqrt(p * (1 - p) * (1/u1 + 1/u2))
    z_stat = (p1-p2)/se
    p = st.norm.cdf(np.abs(z_stat))
    return 2 * (1 - p)
     
```

```python
def result(a, b):
    p_value = z_test(a, b)

    alpha = 0.05

    print('p-значение:', p_value)

    if p_value < alpha:
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")
```

```python
#Создадим переменные для функции
yandex = full_df.query('source == "yandex"')
google = full_df.query('source == "google"')
```

```python
result(yandex, google)
```

Конверсия пользователей в просмотры контактов, пришедших из Yandex равна конверсии пользователей, пришедших из Google.
